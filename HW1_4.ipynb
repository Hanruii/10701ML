```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CSVData = open("/Users/hanrui/Dropbox/Mac (3)/Documents/ML_CMU/10701-HW1/data1.csv")
array = np.genfromtxt(CSVData, delimiter=",")

print(array)
```

```python
sum = np.sum(array)
N = len(array)
theta_MLE = sum/N
print (theta_MLE)
```

```python
sum_n = np.zeros(100)
```

```python
for i in range(len(array)):
    sum_n[i]= np.sum(array[:i+1])
```

```python
Ns = np.arange(1,101,1)
```

```python
theta_n = sum_n/Ns
print(theta_n)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(Ns, theta_n)
plt.savefig('/Users/hanrui/Dropbox/Mac (3)/Documents/ML_CMU/10701-HW1/mle.png')
```

# MAP 4.2

```python
from scipy.stats import beta
fig, ax = plt.subplots(1, 1)
a, b = 1005, 995
x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
ax.plot(x, beta.pdf(x, a, b))
ax.set_title('beta A pdf')
plt.savefig('/Users/hanrui/Dropbox/Mac (3)/Documents/ML_CMU/10701-HW1/beta_A.png')
e_theta_A = beta.stats(a, b, moments='m')
print(e_theta_A)
```

```python
from scipy.stats import beta
fig, ax = plt.subplots(1, 1)
a, b = 20, 16
x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
ax.plot(x, beta.pdf(x, a, b))
ax.set_title('beta B pdf')
plt.savefig('/Users/hanrui/Dropbox/Mac (3)/Documents/ML_CMU/10701-HW1/beta_B.png')
e_theta_B = beta.stats(a, b, moments='m')
print(e_theta_B)
```
